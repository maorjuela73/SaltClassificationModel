```python
from os import listdir

img_names = listdir('./data/train/images/')
img_names.sort()
images = {}
for i in img_names:
    images[i[:-4]] = mpimg.imread('./data/train/images/' + i)[:,:,0]
```

```python
print(images['114d62da56'])
```

```python
msk_names = listdir('./data/train/masks/')
msk_names.sort()
masks = {}

# print(msk_names)
for i in msk_names:
    masks[i[:-4]] = mpimg.imread('./data/train/masks/' + i)
```

```python
print(masks['114d62da56'])
```

```python
implot = plt.imshow(images['114d62da56'], cmap=('gray'))
```

```python
implot = plt.imshow(masks['114d62da56'], cmap=('gray'))
```
